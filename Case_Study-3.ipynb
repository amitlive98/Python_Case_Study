```python
import pandas as pd
terror=pd.read_csv("terror.csv",header=0,encoding="Latin")
terror.head()
```

```python
terror.columns
```

```python
# Q.1)	How many attacks happened in India?
terror[terror.country_txt=='India'].count().iyear
```

```python
# Q.2)	How many attacks happened in India and upto 3 people were killed?
t1=terror[terror.country_txt=='India']
t1[terror.nkill<=3].count().iyear
```

```python
# Q.3)	Extract the city and summary for attacks above
terror[terror.country_txt=='India'][terror.nkill<=3][['city','summary']].dropna(axis=0).head()
```

```python
# Q.4)	In a single terror incident in India, find out top 5 cities by number killed
terror.query("(country_txt=='India')")[["eventid","country_txt","city","nkill"]].sort_values('nkill',ascending=False).head(5)
```

```python
# Q.5)	In a single terror incident in India, find out top 5 cities by number killed and wounded 
t4=terror[terror['country_txt'] == 'India'].groupby('city').agg({'nkill':'sum', 'nwound': 'sum'})
t4['Net'] = t4['nkill'] + t4 ['nwound']
t4.sort_values('Net', ascending = False).head(6)
```

```python
# Q.6) How many attacks were successful that were suicide attacks?
terror.query("(success==1 and suicide==1)").count().iyear
```

```python
# Q.7) label all the incidents where the number killed was more than 5 as severe. 
import numpy as kp
terror['label'] = kp.where(terror['nkill'] > 5, 'severe', 'notsevere')
terror.query("(label=='severe')").count().iyear
```

```python
# #Q.8) write a function to label an incident that was both successful and suicidal
def check(event_no) :
    if ((terror.suicide[terror.eventid == event_no] == 1).tolist()[0] and 
        (terror.success[terror.eventid == event_no] == 1).tolist()[0]) :
        print('Yes,Exist')
    else :
        print('No,Not Exits')
```

```python
check(200707070001)
```

```python
# Q.9)	Create a new category representing if the incident occured in Afghanistan, Pakistan or India 
#as one level of the category and all the other countries as another level
terror['Area']=kp.where(terror['country_txt']==('India' or 'Afghanistan' or 'Pakistan'),'Asia','rest')
terror.head(6)
```

```python
#Q.10)	How many incidents happened in Af-Pak-India vs ROW?
t10=terror.query("(country_txt=='India' or country_txt=='Afghanistan' or country_txt=='Pakistan')").count().iyear
t10
```

```python
# Q.11)	List the number of suicides attacks and average kills by Af-Pak-India vs ROW. Rename columns in the output as 
#Average_Kills and Number_Incidents.
terror[terror.suicide==1].groupby('Area').agg({'suicide':'count','nkill':'mean'})
```
